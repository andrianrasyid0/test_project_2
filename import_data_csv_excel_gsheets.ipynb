# Request API

````python
import requests

# 1. Tentukan URL Endpoint API Cuaca
url = "https://api.open-meteo.com/v1/forecast"

# 2. Siapkan Parameters (Contoh: Kordinat Jakarta)
# Parameter ini mirip seperti yang ada di slide Anda: params = {'key': 'value'}
payload = {
    "latitude": -6.2088,
    "longitude": 106.8456,
    "current_weather": True
}

print("Mengirim HTTP GET request ke server...\n")

# 3. Lakukan Request
response = requests.get(url, params=payload)

# 4. Cek Status Code (Bisa sambil mengingat materi 4xx dan 5xx sebelumnya)
print(f"Status Code: {response.status_code}")

if response.status_code == 200:
    print("Request Sukses! (200 OK)\n")
    
    # 5. Parse JSON response menjadi Dictionary Python
    data = response.json()
    
    # 6. Ekstrak dan tampilkan data yang dibutuhkan
    cuaca_saat_ini = data['current_weather']
    suhu = cuaca_saat_ini['temperature']
    angin = cuaca_saat_ini['windspeed']
    
    print("=== HASIL DATA DARI API ===")
    print(f"Suhu di Jakarta saat ini : {suhu} °C")
    print(f"Kecepatan Angin          : {angin} km/h")
    
else:
    print("Waduh, gagal mengambil data. Coba cek koneksi atau URL-nya.")
````

# Make Your Own Data

### Manual

````python
import pandas as pd

print("=== 1. MANUAL ENTRY ===")
print("Mengetik data satu per satu secara manual...\n")

# Programmer harus mengetik/hardcode setiap nilai data
data_manual = {
    'Nama': ['Budi Santoso', 'Siti Aminah', 'Andi Wijaya'],
    'Umur': [28, 24, 31],
    'Pekerjaan': ['Data Analyst', 'Software Engineer', 'UI/UX Designer']
}

# Mengubah dictionary menjadi format tabel (DataFrame)
df_manual = pd.DataFrame(data_manual)

print(df_manual)
print(f"\nTotal data: {len(df_manual)} baris")
print("-" * 30)
````

### Synthetic

library bernama Faker untuk membuat data buatan (sintetis) yang seolah-olah asli, dan menggunakan looping untuk menggandakannya dalam sekejap.

````python
import pandas as pd
import random
from faker import Faker

print("\n=== 2. SYNTHETIC GENERATION ===")
print("Membuat data palsu secara otomatis dengan AI/Algoritma...\n")

# Inisialisasi Faker dengan lokalisasi Indonesia agar namanya lokal
fake = Faker('id_ID')

data_sintetis = []
jumlah_data_yang_diinginkan = 1000 # Kita bisa ubah ini jadi 10.000 atau 1 juta dengan mudah

# Komputer yang bekerja membuatkan data untuk kita menggunakan looping
for _ in range(jumlah_data_yang_diinginkan):
    baris_baru = {
        'Nama': fake.name(),                     # Generate nama acak
        'Umur': random.randint(22, 55),          # Generate umur acak antara 22-55
        'Pekerjaan': fake.job()                  # Generate nama pekerjaan acak
    }
    data_sintetis.append(baris_baru)

# Mengubah menjadi DataFrame
df_sintetis = pd.DataFrame(data_sintetis)

# Tampilkan 5 data teratas saja
print(df_sintetis.head())
print(f"\nTotal data yang berhasil di-generate: {len(df_sintetis)} baris!")
````
